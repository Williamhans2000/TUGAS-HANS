```python
# Langkah 1: Install dan Import Library
!pip install ipywidgets
import ipywidgets as widgets
from IPython.display import display, clear_output

# Langkah 2: Buat Widget Input
nama = widgets.Text(description='Nama:')
usia = widgets.IntText(description='Usia:')
tanggal_lahir = widgets.DatePicker(description='Tgl Lahir:')
waktu = widgets.Text(description='Waktu:', placeholder='Contoh: 14:30')
jurusan = widgets.Dropdown(
    options=['Informatika', 'Sistem Informasi', 'Teknik Komputer', 'Bisnis Digital'],
    description='Jurusan:'
)
tombol_simpan = widgets.Button(description='Simpan')

# Langkah 3: Fungsi Saat Tombol Diklik
output = widgets.Output()

def simpan_data(b):
    with output:
        clear_output()
        print("=== Data Anda ===")
        print(f"Nama          : {nama.value}")
        print(f"Usia          : {usia.value}")
        print(f"Tanggal Lahir : {tanggal_lahir.value}")
        print(f"Waktu         : {waktu.value}")
        print(f"Jurusan       : {jurusan.value}")

tombol_simpan.on_click(simpan_data)

# Langkah 4: Tampilkan Formulir
display(nama, usia, tanggal_lahir, waktu, jurusan, tombol_simpan, output)
```
