```python
import pandas as pd

# Membaca file CSV
df = pd.read_csv("/nutrition.csv")

# Tampilkan data sebagai TABEL
print("TAB: TABEL")
df.head(10)  # Tampilkan 10 baris pertama
```

```python
# Tampilkan statistik sederhana
print("\nTAB: METRICS")
print(f"Jumlah Makanan: {len(df)}")

if 'calories' in df.columns:
    print(f"Rata-rata Kalori: {df['calories'].mean():.2f}")

if 'fat' in df.columns:
    print(f"Rata-rata Lemak: {df['fat'].mean():.2f}")

if 'protein' in df.columns:
    print(f"Rata-rata Protein: {df['protein'].mean():.2f}")
```

```python
import json

print("\nTAB: JSON")
json_data = df.to_dict(orient='records')
print(json.dumps(json_data[:5], indent=2))  # Tampilkan 5 data pertama
```
