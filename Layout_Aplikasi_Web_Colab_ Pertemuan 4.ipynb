# Aplikasi Web Organisasi
Versi lanjutan: dengan halaman galeri, layanan, dan simpan data kontak ke CSV.

```python
# Simpan aplikasi Streamlit ke file app.py
app_code = '''
import streamlit as st
import pandas as pd
import os

st.set_page_config(page_title="Aplikasi Organisasi", layout="wide")
st.title("Aplikasi Web Organisasi ABC")

menu = st.sidebar.selectbox("Navigasi", ["Beranda", "Tentang Kami", "Layanan", "Galeri", "Kontak"])

if menu == "Beranda":
    st.header("Selamat Datang di Aplikasi Organisasi ABC")
    st.write("Ini adalah halaman utama.")
    with st.container():
        st.subheader("Informasi Umum")
        st.write("Organisasi ABC bergerak di bidang teknologi informasi...")

elif menu == "Tentang Kami":
    st.header("Tentang Kami")
    st.write("""
        Kami adalah organisasi yang berdedikasi untuk menyediakan solusi teknologi bagi masyarakat.
        Didirikan tahun 2020 dan terus berkembang.
    """)

elif menu == "Layanan":
    st.header("Layanan Kami")
    st.write("- Konsultasi Teknologi Informasi")
    st.write("- Pelatihan Digital")
    st.write("- Pengembangan Aplikasi")

elif menu == "Galeri":
    st.header("Galeri Kegiatan")
    st.image([
        "https://via.placeholder.com/300x200",
        "https://via.placeholder.com/300x200"
    ], caption=["Workshop Teknologi", "Pelatihan Masyarakat"], width=300)

elif menu == "Kontak":
    st.header("Hubungi Kami")
    with st.form(key='contact_form'):
        nama = st.text_input("Nama Lengkap")
        email = st.text_input("Email")
        pesan = st.text_area("Pesan")
        submit_button = st.form_submit_button(label="Kirim")

        if submit_button:
            # Simpan ke CSV
            data = {'Nama': [nama], 'Email': [email], 'Pesan': [pesan]}
            df = pd.DataFrame(data)
            if not os.path.isfile("data_kontak.csv"):
                df.to_csv("data_kontak.csv", index=False)
            else:
                df.to_csv("data_kontak.csv", mode='a', header=False, index=False)
            st.success(f"Terima kasih, {nama}. Pesanmu telah dikirim!")
'''

with open("app.py", "w") as f:
    f.write(app_code)
```

```python
# Jalankan Streamlit dan buka akses dengan localtunnel
import os
import threading

def run():
    os.system('streamlit run app.py &')

threading.Thread(target=run).start()
```
